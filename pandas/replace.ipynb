```python
import seaborn as sns
import pandas as pd
```

### map
辞書形式で指定した置き換え

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
data["sex"] = data["sex"].map({"male":0, "female":1})
data["sex"].unique()
```

### apply
関数を適用した置き換え

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
def age_range(age):
    if age < 12:
        return "low_age1"
    if age < 20:
        return "low_age2"
    if age < 40:
        return "middle_age1"
    if age < 60:
        return "middle_age2"
    else:
        return "high_age"

data["age"] = data["age"].apply(age_range)
data["age"].unique()
```

### applymap
要素毎にapply

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
data = data.applymap(lambda x: 1 if (x == "man" or x == "male") else x)
data.head()
```
