#### データサイエンティストの要件
_"まずは、分析スピード。仮説を考えて、データで確認するというプロセスをいかに高速に回せるか。短い時間でたくさんのアイデアを思いつき、データで検証することができれば、発見の確率が上がります。データをスピーディーに分析するためには、コンピューターの基本的な知識と、プログラミングスキル、数学的なスキル、データ解釈スキルが必要。"_  
引用：[AIを駆使する最先端で高収入な「データサイエンティスト」になるためには？](https://gigazine.net/news/20171028-udemy-data-science/)

データサイエンティスト協会が出している「データサイエンティストに求められるスキルセット」は下図の通りです。  
![s01_m.png](https://i.gzn.jp/img/2017/10/28/udemy-data-science/s01_m.png)
  
３つのスキルをどう育てていくかはバックグランドによります。例えば、元々エンジニアの人は「データエンジニアリング力」が現在位置になるかと思います。そこから、データエンジニアリング部分を補いつつ、時計回りか反時計回りで学習を進めるか、あるいは2つのベクトルを同時に進めていくか、というイメージです。なお、エンジニアとしてある程度の経験（オブジェクト指向設計、デバッグ、アルゴリズム構築）を積んでいる方の「データエンジニアリング力」は既に高い位置にあるはずです。学習コストはそこまで高くありません。個人的な見解ですが、既にビジネスサイドにいて、1つの武器としてデータサイエンスを用いるような場合は、「データサイエンス力」＝70%、「データエンジニアリング力」＝30%というような割合で学習を進めるとよいような気がします。多少は手を動かして実装してみて理解を深めると良い気がします。

また、製品開発においてAIを組み込んでいく場合は、狭義には「機械学習エンジニア」という方が世間的には正しいかもしれません。

#### 特徴量と目的変数
* 特徴量
各データを表現する変数を「特徴量」と呼びます。  
統計学では「説明変数」と呼ぶこともあります。

* 目的変数
各データに対する出力を「目的変数」や「正解データ」、「教師データ」などと呼びます。  
統計学では「被説明変数」と呼ぶこともあります。

教師あり学習は「特徴量」＋「目的変数」をセットを持ち、教師なし学習は「特徴量」のみを持ちます。これが両者の大きな違いです。

|学習アルゴリズム|データセット|
|--|--|
|教師あり|特徴量, 目的変数|
|教師なし|特徴量|


#### 回帰と分類
##### 回帰

連続値（上記で言う量的データ）を予測します。  
特徴量による目的変数の変化を何かしらの式にフィットさせます。

```python
X = np.arange(50)
f = lambda x: 0.2 * x -3
y = f(X) + np.random.randn(50)
plt.scatter(X, y)
plt.plot(X, f(X), c="red")
```

* 分類
離散値（上記で言う質的データ）を予測します。  
特徴量により分布する目的変数を分割するような線を引くイメージです。

```python
X = np.random.randn(100); y = np.random.randn(100)
f = lambda x: 0.5 * x -0.5
labels = f(X)>y
plt.scatter(X, y, c=labels)
plt.plot(X, f(X), c="red")
```

### 教師あり学習を用いて予測モデルを作成してみる
* KNNを用いる。  
https://qiita.com/yshi12/items/26771139672d40a0be32

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split # trainとtest分割用

# train用とtest用のデータ用意。test_sizeでテスト用データの割合を指定。random_stateはseed値を適当にセット。
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=3) 
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6) # インスタンス生成。n_neighbors:Kの数
knn.fit(X_train, y_train)                 # モデル作成実行
```

```python
knn.score(X_test, y_test)
```

```python
y_pred = knn.predict(X_test)              # 予測実行
```

```python
y_test.values
```

```python
y_pred
```
