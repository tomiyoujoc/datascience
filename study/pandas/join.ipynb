```python
import seaborn as sns
import pandas as pd
```

### merge
howに"inner","right","left","outer"を指定することで結合方式を選択

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
embarked = pd.DataFrame({"letter":["S", "C", "Q"], "town":data["embark_town"].dropna().unique().tolist()})
embarked.head()
```

```python
del data["embark_town"]
data.head()
```

```python

data = pd.merge(data, embarked, left_on="embarked", right_on="letter", how="inner")
data.head()
```

### rolling
窓関数を適用する

```python
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
flights["passengers"].rolling(center=False, window=3, min_periods=1).sum().head()
```

```python
#窓の中心に結果の値を格納する
flights["passengers"].rolling(center=True, window=3, min_periods=1).sum().head()
```

### cross join
* 全結合用の関数はないので自前で処理する

```python
a = pd.DataFrame([1, 2, 3], columns=["a"])
a
```

```python
b = pd.DataFrame([4, 5, 6], columns=["b"])
b
```

```python
# 結合用に全て同じ値のキーを用意
a["key"] = 0
b["key"] = 0
```

```python
ab = pd.merge(a, b, on="key")
ab.drop("key", axis=1)
```
